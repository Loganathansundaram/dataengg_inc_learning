```python
df = spark.read.format("csv").load("/Volumes/bricks_project/ecommerce/datalake/olist_products_dataset.csv",header=True,inferSchema=True)
df1 = df.limit(1000)
```

```python
# drop if exists bricks_project.deltadb
spark.sql("drop table if exists bricks_project.deltadb")
# Create the database if it doesn't exist
spark.sql("CREATE DATABASE IF NOT EXISTS bricks_project.deltadb")
# write the table
df1.write.saveAsTable("bricks_project.deltadb.products", mode='overwrite')

```
