```python
%py
import requests

# Step 1: API call to fetch CSV content
url = "https://public.tableau.com/app/sample-data/mobile_os_usage.csv"
response = requests.get(url)

# File path
file_path = "/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv"

# Step 2: Check if file exists
file_list = [f.name for f in dbutils.fs.ls("/Volumes/workspace/default/usage_metrics/")]

if "mobile_os_usage.csv" not in file_list:
    dbutils.fs.put(file_path, response.text, overwrite=False)
    print("File created successfully.")
else:
    print("File already exists â€” skipping write.")
```

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW mobile_usage
USING CSV
OPTIONS (
  path "/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",
  header "true",
  inferSchema "true"
);
```
